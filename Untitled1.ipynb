```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

```

```python
df=pd.read_csv('IceCreamData.csv')
```

```python
x_train, y_train = df.iloc[:490, 0].values, df.iloc[:490, 1].values
x_test, y_test = df.iloc[490:500, 0].values, df.iloc[490:500, 1].values
```

```python
w, b = np.random.randn(), np.random.randn()
learning_rate, epochs = 0.0001, 1000
m = len(y_train)
```

```python
def compute_cost(y_true, y_pred):
    return np.sum((y_true - y_pred) ** 2) / len(y_true)
```

```python
cost_history, w_history, b_history = [], [], []
for _ in range(epochs):
    y_pred = w * x_train + b
    dw = (-2/m) * np.sum(x_train * (y_train - y_pred))
    db = (-2/m) * np.sum(y_train - y_pred)
    
    w -= learning_rate * dw
    b -= learning_rate * db
    
    cost = compute_cost(y_train, y_pred)
    cost_history.append(cost)
    w_history.append(w)
    b_history.append(b)
```

```python
df_updates = pd.DataFrame({
    "Weight": w_history,
    "Bias": b_history,
    "Cost": cost_history
})
df_updates.to_csv("IceCreamData.csv", index=False)
```

```python
y_test_pred = w * x_test + b
```

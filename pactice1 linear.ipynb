```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Load Data
data = pd.read_csv("IceCreamData.csv")
X = data["Weight"].values
Y = data["Cost"].values
bias = data["Bias"].values  # Given bias

# Normalize X and Y to prevent exploding gradients
X = (X - np.mean(X)) / np.std(X)
Y = (Y - np.mean(Y)) / np.std(Y)

# Split data into training and testing
split = int(0.8 * len(X))
X_train, Y_train = X[:split], Y[:split]
X_test, Y_test = X[split:], Y[split:]

# Initialize parameters
m = 0  # Initial slope
learning_rate = 0.001  # Reduce learning rate to prevent large updates
epochs = 100

# Gradient Descent
for _ in range(epochs):
    Y_pred = m * X_train + bias[:split]  # Use given bias
    error = Y_train - Y_pred
    mse = np.mean(error**2)  # Mean Squared Error

    # Compute gradient w.r.t m
    dm = -2 * np.mean(X_train * error)

    # Update m (bias is already given)
    m = m - learning_rate * dm

# Print the final equation
print(f"Final Equation: Y = {m:.4f}X + Given Bias")

# Function to compute loss (MSE)
def compute_loss(X, Y, m, bias):
    Y_pred = m * X + bias
    return np.mean((Y - Y_pred) ** 2)

# Compute loss for training and testing data
train_loss = compute_loss(X_train, Y_train, m, bias[:split])
test_loss = compute_loss(X_test, Y_test, m, bias[split:])

print(f"Training Loss: {train_loss:.4f}")
print(f"Testing Loss: {test_loss:.4f}")

# Plot Data and Regression Line
plt.scatter(X, Y, color="red", label="Actual Data")
plt.plot(X, m * X + bias, color="blue", label="Best Fit Line")
plt.xlabel("Weight (Normalized)")
plt.ylabel("Cost (Normalized)")
plt.title("Linear Regression - Ice Cream Data (Using Given Bias)")
plt.legend()
plt.show()
```
